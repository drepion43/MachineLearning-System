## 1. K-means를 이용한 이미지 군집화

```python
import cv2
import sys
import numpy as np
from matplotlib import pyplot as plt

def image_kmeans(fin_img,K,fout_img):
    img=cv2.imread(fin_img)
    
    Z=img.reshape((-1,3))
    
    Z=np.float32(Z)
    
    criteria=(cv2.TERM_CRITERIA_EPS+cv2.TERM_CRITERIA_MAX_ITER,10,1.0)
    n_init=10
    ret,label,center=cv2.kmeans(Z,K,None,criteria,n_init,cv2.KMEANS_RANDOM_CENTERS)
    
    center=np.uint8(center)
    
    res=center[label.flatten()]
    res2=res.reshape((img.shape))
    
    plt.imshow(cv2.cvtColor(res2,cv2.COLOR_BGR2RGB))
    plt.title('k=%d',K)
    plt.savefig(fout_img)
```

## 2.PCA를 이용한 사람얼굴 인식

```python
https://scikit-learn.org/stable/auto_examples/applications/plot_face_recognition.html#sphx-glr-auto-examples-applications-plot-face-recognition-py
```

```python
import cv2
import numpy as np
import sys
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.decomposition import PCA
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
from sklearn.datasets import fetch_lfw_people
from sklearn.utils.fixes import loguniform
```

```python
def read_data(fin):
    target_li=[]
    data_li=[]
    
    for line in open(fin):
        working_dir=fin.split('/')[:-1]
        img_path,face_id=line.strip().split(';')
        img_path='/'.join(working_dir)+'/'+img_path
        target_name=img_path.split('/')[1]
        img_data=cv2.imread(img_path,cv2.COLOR_BGR2GRAY)
        
        data_li.append(img_data)
        target_li.append(int(face_id))
        
    return (np.array(data_li),np.array(target_li))
```

```python
def extract_features(x_train,x_test,n_comp):
    pca=PCA(n_components=n_comp,svd_solver='randomized',whiten=True).fit(x_train)
    
    x_train_pca=pca.transform(x_train)
    x_test_pca=pca.transform(x_test)
    
    return x_train_pca,x_test_pca

def train_test_classifer(x_train_pca,x_test_pca,y_train,y_test):
    param_grid = {"C": loguniform(1e3, 1e5),"gamma": loguniform(1e-4, 1e-1),}
    
    clf = RandomizedSearchCV(SVC(kernel="rbf", class_weight="balanced"), param_grid, n_iter=10)
    clf=clf.fit(x_train_pca,y_train)
    
    print("Best estimator found by grid search:")
    print(clf.best_estimator_)
    
    print('predictin people name')
    y_pred=clf.predict(x_test_pca)
    print(classification_report(y_test,y_pred))
```

```python
lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)
```

```python
n_samples, h, w = lfw_people.images.shape
X=lfw_people.data
n_features = X.shape[1]
y = lfw_people.target
target_names = lfw_people.target_names
n_classes = target_names.shape[0]

print("Total dataset size:")
print("n_samples: %d" % n_samples)
print("n_features: %d" % n_features)
print("n_classes: %d" % n_classes)
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
print(x_train.shape)

n_eigenface=150
x_train_pca,x_test_pca=extract_features(x_train,x_test,n_eigenface)
train_test_classifer(x_train_pca,x_test_pca,y_train,y_test)
```

```python
def extract_features(x_train,x_test,n_comp,h,w):
    pca=PCA(n_components=n_comp,svd_solver='randomized',whiten=True).fit(x_train)
    eigenfaces=pca.components_.reshape((n_comp,h,w))
    x_train_pca=pca.transform(x_train)
    x_test_pca=pca.transform(x_test)
    
    return x_train_pca,x_test_pca,eigenfaces
```

```python
from matplotlib import pyplot as plt

def plot_gallery(img,n_col=5):
    
    n_row=round(img.shape[0]/n_col)
    plt.figure(figsize=(1.8*n_col,2.4*n_row))
    plt.subplots_adjust(bottom=0.1,left=0.01,right=0.99,top=0.90,hspace=0.35)
    
    for i in range(n_row*n_col):
        plt.subplot(n_row,n_col,i+1)
        
        plt.imshow(img[i],cmap='gray')
        plt.xticks(())
        plt.yticks(())
```

```python
n_eigenface=150
x_train_pca,x_test_pca,eigenface=extract_features(x_train,x_test,n_eigenface,h,w)
train_test_classifer(x_train_pca,x_test_pca,y_train,y_test)
plot_gallery(eigenface)
plt.show()
```
